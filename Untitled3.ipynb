```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[["TARGET VARIABLE","EMEA, GDP Normalized","EMEA, Employment Rate"]])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]

plt.scatter(df0['TARGET VARIABLE'],df0['EMEA, GDP Normalized'],df0["EMEA, Employment Rate"],color='green')
plt.scatter(df1['TARGET VARIABLE'],df1['EMEA, GDP Normalized'],df1["EMEA, Employment Rate"],color='red')
plt.scatter(df2['TARGET VARIABLE'],df2['EMEA, GDP Normalized'],df2["EMEA, Employment Rate"],color='black')

plt.xlabel('TARGET VARIABLE')
plt.ylabel('EMEA, GDP Normalized')
plt.legend()
```

```python
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df2)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
